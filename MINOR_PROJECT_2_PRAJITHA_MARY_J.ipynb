```python
# Count duplicate rows

print("DUPLICATE ROWS:",df.duplicated().sum())

# Remove duplicates

df = df.drop_duplicates()
print("AFTER REMOVING DUPLICATES:",df.shape)
```

```python
# Convert Date column into datetime format

df["Date"] = pd.to_datetime(
    df["Date"],
    errors="coerce",
    dayfirst=True
)

# invalid dates

print("INVALID DATES:", df["Date"].isnull().sum())
```

```python
# Convert Amount column into string

df["Amount"] = df["Amount"].astype(str)

# Remove Rupee Symbol

df["Amount"] = df["Amount"].str.replace("₹","", regex=False)

# Remove Rs.

df["Amount"] = df["Amount"].str.replace("Rs.","", regex=False)

# Remove commas

df["Amount"] = df["Amount"].str.replace(",","", regex=False)

# Remove extra spaces

df["Amount"] = df["Amount"].str.strip()

# Convert into numeric

df["Amount"] = pd.to_numeric(
    df["Amount"],
    errors="coerce"
)

print("INVALID AMOUNTS:", df["Amount"].isnull().sum())
```

```python
# Convert Type column into lowercase

df["Type"] = df["Type"].str.lower()

# Replace values

df["Type"] = df["Type"].replace({
    "dr":"debit",
    "cr":"credit"
})

print(df["Type"].value_counts())
```

```python
# Replace empty values with NaN

df["Mode"] = df["Mode"].replace("", np.nan)

# Check missing values

print("MISSING VALUE:",df["Mode"].isnull().sum())
```

```python
# Remove rows having invalid Date or Amount

df = df.dropna(subset=["Date","Amount"])
print("DATASET SHAPE:",df.shape)
```

```python
# Function to identify vendor

def extract_vendor(description):
    description = str(description).upper()
    for vendor in vendor_dict:
        keywords = vendor_dict[vendor]
        for word in keywords:
            if word in description:
                return vendor

    return "Uncategorised"
```

```python
# Extract hour from Time column

df["hour"] = df["Time"].str[:2].astype(int)

# Display first few rows

print(df[["Time", "hour"]].head())
```

```python
# Function to classify time of the day

def time_period(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"


# Apply function

df["time_period"] = df["hour"].apply(time_period)

print(df[["Time", "hour", "time_period"]].head())
```

```python
# Create a new dataframe with only debit transactions

anomaly_df = df[df["Type"] == "debit"].copy()
```

```python
# Calculate average amount for each category

category_mean = anomaly_df.groupby("category")["Amount"].transform("mean")

anomaly_df["category_mean"] = category_mean

print(anomaly_df[["category", "Amount", "category_mean"]].head())
```

```python
# Calculate standard deviation for each category

category_std = anomaly_df.groupby("category")["Amount"].transform("std")

anomaly_df["category_std"] = category_std

print(anomaly_df[["category", "category_std"]].head())
```

```python
# Calculate Z-score

anomaly_df["z_score"] = (

    anomaly_df["Amount"] -

    anomaly_df["category_mean"]

) / anomaly_df["category_std"]

print(anomaly_df[["Amount", "z_score"]].head())
```

```python
# Replace infinity values

anomaly_df["z_score"] = anomaly_df["z_score"].replace([np.inf, -np.inf], np.nan)

# Replace NaN values with 0

anomaly_df["z_score"] = anomaly_df["z_score"].fillna(0)
```

```python
# Transactions having Z-score greater than 2

anomalies = anomaly_df[anomaly_df["z_score"] > 2]

print("NUMBER OF ANOMALIES")

print(len(anomalies))
```

```python
# Sort by Z-score

anomalies = anomalies.sort_values(

    by="z_score",

    ascending=False

)

print(anomalies.head())
```
